 Importing the Libraries 

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image
from IPython.display import display
plt.style.use('seaborn')
```

Reading the input image

Now we will be loading the image file into our program by making use of the imread function which takes the path of the file. 
We will also be changing the color code of the image to RGB format to get the original colors and plot the image using the imshow function.

```python
img = cv2.imread("input_image.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(img)
plt.show()
```

Converting Orignal to grey image

To decrease the complexity of the image and make the handling easier we will be converting the image to a grayscale image with the help of the cvtColor function. The same is plotted with the help of the code mentioned below.

```python
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.axis('off')
plt.figure(figsize=(8,8))
plt.imshow(gray_img,cmap="gray")
plt.title("GrayScale Image")
plt.show()
```

Inverting the grey image

Now the next step is to invert the image,In order to do the same is that when a image is inverted it helps in studying the image with more precision and detail.
The more detailed study is important for the sketch generation as the sketches need to be accurate and good. The code to invert the image is shown below.


Image: Invert formula 


```python
#inv_gray_img = 255 - gray_img
inv_gray_img = cv2.bitwise_not(gray_img)
plt.axis('off')
plt.imshow(inv_gray_img)
plt.show()
```

Blurring the inverted grey image

GuassianBlur -> In addition to this, we will also be smoothing the image to make sure the sketch we obtain is less edgy and smooth. The code for the same is shown below.


Image: Gaussian blur function



```python
blur_img = cv2.GaussianBlur(inv_gray_img,(21,21),0)
plt.axis('off')
plt.imshow(blur_img)
plt.show()
```

Inverting the blurred image

```python
#inv_blur_img = 255 - blur_img
inv_blur_img = cv2.bitwise_not(blur_img)
plt.axis('off')
plt.imshow(inv_blur_img)
plt.show()
```

Creating the pencil sketch image

Now that the whole image processing is done, we will be passing the previous outputs to the function and pass the right and accurate parameters to make the appropriate changes to the image. The same is mentioned in the code below.

```python
pencil_img = cv2.divide(gray_img,inv_blur_img,scale = 256.0)
plt.axis('off')
plt.imshow(pencil_img)
plt.show()
```

```python
cv2.imwrite("./output_image.png",pencil_img)
```

```python
plt.figure(figsize=(20,20))
plt.subplot(1,5,1)
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.subplot(1,5,2)
plt.imshow(gray_img,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.subplot(1,5,3)
plt.imshow(inv_gray_img,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.subplot(1,5,4)
plt.imshow(inv_blur_img,cmap="gray")
plt.axis("off")
plt.title("Smoothen Image")
plt.subplot(1,5,5)
plt.imshow(pencil_img,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```
